#**MODIFYING GRAPH**
plot graph->plt.plot(arr_1,arr_2)
->**including linestyle**
plt.plot(arr_1,arr_2,line style="style name")
->**to get graph with grid lines**
plt.grid()
->**to label the axis**
plt.xlabel("label")
plt.ylabel("label")
->**to give graph an title**
plt.title("title")

```python
plt.plot(x,y, linestyle="dotted",c="g")
plt.grid()
plt.xlabel("THis is x axis")
plt.ylabel("This is y axis")
plt.title("MATPLOT USAGE EXAMPLE WITH LINE GRAPH")
plt.show()
```

#**SUB PLOT**
multiple plots in a image
plt.subplot(row,column,figure no)
In below example
(1,2,1)
1,2 is common in both the plots->represents that the plots are arranged in a image->int he given image the plots are aranged in 1 row tw columns

```python
a=np.array([1,2,3,4])
b=np.array([5,6,7,8])

plt.subplot(1,2,1)
plt.plot(a,b)

x=np.array([1,2,3,4])
y=np.array([20,50,60,150])

plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
```

```python
c=np.array([2,4,6,8])
d=np.array([10,30,6,1])

plt.subplot(2,1,1)
plt.plot(c,d)

e=np.array([3,6,9,12])
f=np.array([4,8,12,16])

plt.subplot(2,1,2)
plt.plot(e,f)

plt.show()
```
